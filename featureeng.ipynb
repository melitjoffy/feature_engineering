```python
import pandas as pd
import numpy as np
```

```python
data={
    'Age':[25,np.nan,30,22,np.nan],
    'Salary':[50000,60000,np.nan,52000,58000],
    'Department':['IT','HR','Finance',np.nan,'Finance']
}
df=pd.DataFrame(data)
df
```

imputation:replace missing numerical values with mean

```python
from sklearn.impute import SimpleImputer
```

mpute with mean value

```python
mean_imputer=SimpleImputer(strategy='mean')

```

```python
df[['Age']]=mean_imputer.fit_transform(df[['Age']])
```

```python
df['Age']
```

```python
median_imputer=SimpleImputer(strategy='median')
```

```python
df[['Salary']]=median_imputer.fit_transform(df[['Salary']])
```

```python
df['Salary']
```

```python

from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python

from sklearn.feature_selection import f_classif
```

```python

from sklearn.datasets import load_iris
```

```python

X,y=load_iris(return_X_y=True,as_frame=True)
```

```python
print("Original feature names:",X.columns.tolist())
```

```python
y
```

```python
selector=SelectKBest(score_func=f_classif,k=2)
X_new=selector.fit_transform(X,y)
#f_classif=annova
#k=no. of features to select

```

```python
selector=SelectKBest(score_func=mutual_info_classif,k=2)
X_new=selector.fit_transform(X,y)
#f_classif=annov
```

```python
selected_features=X.columns[selector.get_support()]
selected_features.tolist()
```
